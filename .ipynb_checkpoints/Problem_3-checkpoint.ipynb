Problem 3:
1. Choose a random non-zero 2-vector z in R^2, and create some data y to be a noisy estimate of the 2-vector such that y = z + epsilon, where epsilon is draw from a 2d normal with zero mean and variance tensor: V = sigma^2 I_2.
2. Our goal is to estimate z, given y, and knowing sigma.
3. Compute the likelihood function p(y|z).
4. Notice that z can be parametrized in terms of r and phi, and the likelihood can be written as p(y|r,phi).
5. Find the average of the likelihood using the methods in problem 1 and 2. 

```python
import numpy as np
import random as random
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from itertools import product as iproduct
from scipy.integrate import quad
```

Generating our random vector, error, and data. 

```python
rng = np.random.default_rng(seed=42)
z = rng.standard_normal(2)
z_uc = np.linalg.norm(z)
z_mag = rng.uniform(0.5, 3.0)

sigma = 0.5
epsilon = rng.multivariate_normal(mean=[0,0], cov=sigma**2 * np.eye(2))

y = z + epsilon

print(z, epsilon, y)
```

```python
def likelihood(y, r, phi, sigma):
    z = r * np.array([np.cos(phi), np.sin(phi)])
    residual = y - z
    exponent = -0.5 * residual @ residual /sigma**2
    norm = 1.0 / (2 * np.pi * sigma**2)
    return norm * np.exp(exponent)
```

Computing the log liklihood p(y|z).

```python
def log_likelihood(y, z, sigma):
    mean = z
    cov = sigma**2 * np.eye(2)
    return multivariate_normal.logpdf(y, mean=mean, cov=cov)

print(f"log p(y|z) = {log_likelihood(y, z, sigma):.6f}")
```
