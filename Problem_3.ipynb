```python
import numpy as np
import random as random
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from itertools import product as iproduct
from scipy.integrate import quad
```

```python
rng = np.random.default_rng(seed=42)
z = rng.uniform(low=-5, high=5, size=2)
while np.allclose(z,0):
    z = rng.uniform(low=-5, high=5, size=2)

sigma = 1.0
epsilon = rng.multivariate_normal(mean=[0,0], cov=sigma**2 * np.eye(2))

y = z + epsilon

print(z, epsilon, y)
```

```python
def log_likelihood(y, z, sigma):
    mean = z
    cov = sigma**2 * np.eye(2)
    return multivariate_normal.logpdf(y, mean=mean, cov=cov)

print(f"log p(y|z) = {log_likelihood(y, z, sigma):.6f}")
```
